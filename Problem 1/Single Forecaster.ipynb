## Import The Training Dataset

```python
import pandas as pd
import numpy as np
```

```python
historical_weather = pd.read_csv('DataSet/historical_weather.csv')
historical_weather
```

```python
historical_weather = historical_weather.drop(
    columns=["precipitation_mm", "snow_depth_mm", "avg_wind_dir_deg", "avg_wind_speed_kmh", "max_temp_c", "min_temp_c"])
historical_weather
```

```python
historical_weather = historical_weather.dropna(subset=["avg_temp_c"])
historical_weather
```

```python
historical_weather.dtypes
```

Convert the date to a day counted from 2014-01-01

```python
#Convert to date time
historical_weather["date"] = pd.to_datetime(historical_weather["date"])
# Reference date
reference_date = pd.to_datetime("2014-01-01")
#Calculate the day and replace
historical_weather["date"] = (historical_weather["date"] - reference_date).dt.days
historical_weather
```

```python
historical_weather["city_id"] = historical_weather["city_id"].str[1:].astype(int)
historical_weather
```

```python
rows_count = len(historical_weather["date"].unique())
```

Create the multi-output form

```python
# Create an empty dataframe
df = pd.DataFrame(index=range(rows_count))
# Assign 'date' column values from 0 to 1825
df['date'] = range(rows_count)

# List of column names to add
col_names = historical_weather["city_id"].unique()

# Add the additional columns with NaN values
for col in col_names:
    df[col] = np.nan

df
```

```python
df.drop(columns=["date"], inplace=True)
df
```

```python
# Iterate through each cell in the DataFrame
for i, row in df.iterrows():
    for j, value in row.items():
        # Row index = i, Column = j , cell value = value
        filtered_df = historical_weather[(historical_weather['city_id'] == j) & (historical_weather['date'] == i)]
        if not filtered_df.empty:
            df.at[i, j] = value = filtered_df['avg_temp_c'].values[0]
df
```

```python
# Replace NaN values with column averages
for col in df.columns:
    col_mean = df[col].mean()
    df[col].fillna(col_mean, inplace=True)

df
```

```python
# Add an index column named 'date'
df = df.reset_index().rename(columns={'index': 'date'})
df
```

```python
# Specify the file path where you want to save the CSV file
file_path = 'Data_processed.csv'
# Save the DataFrame to CSV
df.to_csv(file_path, index=False)
```

Normalize the data

```python
# Calculate mean and standard deviation for each column
# Calculate mean and standard deviation for each column (excluding the first column)
column_stats = [
    {
        'column': col,
        'mean': df[col].mean(),
        'std': df[col].std()
    }
    for col in df.columns[1:]  # Exclude the first column
]
column_stats
```

```python
# Normalize the DataFrame
normalized_df = df.copy()
for col_stat in column_stats:
    col = col_stat['column']
    mean_val = col_stat['mean']
    std_val = col_stat['std']
    normalized_df[col] = (normalized_df[col] - mean_val) / std_val

normalized_df
```

```python
# Specify the file path where you want to save the CSV file
file_path = 'normalized_df.csv'
# Save the DataFrame to CSV
normalized_df.to_csv(file_path, index=False)
```

## Tensorflow LSTM

```python
import  tensorflow as tf
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense ,LSTM
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# split the sequence into samples
def split_sequences(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
        end_index = i + n_steps
        if end_index > len(sequences) - 1:
            break
        seq_x, seq_y = sequences[i:end_index], sequences[end_index]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)

def create_multivariate_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        # Extract sequence of length 'sequence_length' and the target at the end
        seq = data[i:i + sequence_length]
        target = data[i + sequence_length]
        X.append(seq)
        y.append(target)
    return np.array(X), np.array(y)

def rmse(y_true, y_pred):
    return tf.sqrt(tf.reduce_mean(tf.square(y_true - y_pred)))
```

```python
for column_name in df.columns:
    new_df = df[column_name].copy()
    n_steps = 14
    
    X,y = split_sequences(new_df, n_steps)
    
    n_features  = 1
    
    X = X.reshape((X.shape[0], X.shape[1], n_features))
    
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
    model.add(Dense(1))  # One Value
    model.compile(loss=rmse, optimizer='adam')
    history = model.fit(X, y, epochs=12, batch_size=5)

    predictions_list=[]

    #------------------Loop for 7 times ---------
    # Get the last 5 values of column '1' as a list
    for i in range(7):
        input_list = new_df.iloc[-n_steps:].tolist()
        #Predictions
        x_input = array(input_list)  #11.8
        x_input = x_input.reshape((1,n_steps, n_features))
        y_pred = model.predict(x_input)
        value = float(y_pred[0][0])
        predictions_list.append(value)
        #Add a new row with a specified value using the `loc` method
        new_df.loc[len(new_df)] = value

    final_df[column_name].iloc[-len(predictions_list):] = predictions_list
    
```

```python
date_index = final_df.index
# De-normalize the DataFrame
denormalized_data = scaler.inverse_transform(final_df)

# Create a new DataFrame with de-normalized data and original date index
denormalized_df = pd.DataFrame(denormalized_data, index=date_index, columns=df.columns)
denormalized_df
```
